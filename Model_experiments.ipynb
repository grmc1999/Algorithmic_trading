```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
Tlist=get_tickers(page=3,vectorized=True,options=[Screener.SectorOption.BASIC_MATERIALS,Screener.IndexOption.SANDP_500])
Tlist=np.unique(np.array(Tlist)).tolist()
data = yf.download(Tlist[:], period = "1y",interval = "1h")
```

```python
train_score=[]
test_score=[]
for T in Tlist:
    X,Y=past_data_prep(data,window=3,target_ticker=T)
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y>0, test_size=0.33,shuffle=False)
    clf=make_pipeline(StandardScaler(),PCA(n_components=5,SVC(kernel="linear"))
    clf.fit(X_train,Y_train)
    tr_sc=clf.score(X_train,Y_train)
    te_sc=clf.score(X_test,Y_test)
    train_score.append(tr_sc)
    test_score.append(te_sc)

print(max(test_score))

import matplotlib.pyplot as plt

train_pos = np.arange(len(Tlist))

fig, ax = plt.subplots(figsize=(30,10))

hbars = ax.barh(train_pos-0.2,train_score,0.4, align='center',label="train")
hbars = ax.barh(train_pos+0.2,test_score,0.4, align='center',label="test")

plt.yticks(train_pos, Tlist)
plt.xticks(np.arange(0,1,0.05))

# label should have dates and model details
label=str(list(data.index)[0].date())+"-"+str(list(data.index)[-1].date())

plt.title(label)

plt.legend()
plt.show()
fig.savefig("SVC"+label+'.png')
```
