```python
def compute_kld(data,dist):
    s_dist=sample(dist,size=data.shape)
    return np.sum(data*np.log(data/s_dist))

def basic_seasonal_analysis(data,ticker,days,value="Close"):
    print(data)
    X=copy.deepcopy(data[value][ticker])
    if 24*days*2<X.dropna().values.shape[0]:
        sd=seasonal_decompose(x=X.dropna(),
                  model="additive",
                   period=int(24*days)
                  )

        #Uniform score
        US=np.std(sd.trend.dropna())
        #residuals score
        #prob count
        c,v=np.histogram(sd.resid.dropna().values/(np.max(np.abs(sd.resid.dropna().values))),bins=60)
        prob=c/sd.resid.dropna().values.shape[0]
        max_v=np.max(np.abs(sd.resid.dropna().values))
        return sd,prob,max_v,US
    else:
        return None,None,None,None
    
class multi_market(object):
    def __init__(self,data):
        self.data=data
        self.v_basic_seasonal_analysis=np.vectorize(self.basic_seasonal_analysis,signature="(),()->(),(j),(),()")
    def basic_seasonal_analysis(self,ticker,days,value="Close"):
        X=copy.deepcopy(self.data[value][ticker])
        if 24*days*2<X.dropna().values.shape[0]:
            sd=seasonal_decompose(x=X.dropna(),
                  model="additive",
                   period=int(24*days)
                  )

            #Uniform score
            US=np.std(sd.trend.dropna().values/(np.max(sd.trend.dropna().values)))
            #residuals score
            #prob count
            c,v=np.histogram(sd.resid.dropna().values/(np.max(np.abs(sd.resid.dropna().values))),bins=60)
            prob=c/sd.resid.dropna().values.shape[0]
            max_v=np.max(np.abs(sd.resid.dropna().values))
            return sd,prob,max_v,US
        else:
            return None,None,None,None
```
